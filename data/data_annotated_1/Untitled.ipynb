```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

im = Image.open('bracketsample_000184.jpg')

# Create figure and axes
fig, ax = plt.subplots()

# Display the image
ax.imshow(im)

# Create a Rectangle patch
rect = patches.Rectangle((0.256010*416, 0.608173*416), 0.257212*416, 0.533654*416, linewidth=0.5, edgecolor='r', facecolor='none')

# Add the patch to the Axes
ax.add_patch(rect)

plt.show()
```
